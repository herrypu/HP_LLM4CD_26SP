```python
def ask_deepseek(prompt):
    from openai import OpenAI
    client = OpenAI(
        api_key = "",
        base_url = "https://api.deepseek.com"
    )
    completion = client.chat.completions.create(
        model="deepseek-reasoner",
        messages=[{"role": "user", "content": prompt}],
        max_tokens=2048,
        temperature=0.7
    )
    content = completion.choices[0].message.content

    print("-" * 30 + " LLM Response " + "-" * 30)
    print(content)
    print("-" * 74)

    return content
```

```python
verilog_generation_prompt = '''
I am trying to create a Verilog model binary_to_bcd_converter for a binary to binary-coded-decimal converter. It must meet the following specifications:
	- Inputs:
		- Binary input (5-bits)
	- Outputs:
		- BCD (8-bits: 4-bits for the 10's place and 4-bits for the 1's place)

How would I write a design that meets these specifications?
'''
```

```python
ask_deepseek(verilog_generation_prompt)
```

### 3. Extraction and Verification

```python
output_verilog_code = '''
module binary_to_bcd_converter (
    input [4:0] binary_in,    // 5-bit binary input (0-31)
    output reg [7:0] bcd_out   // 8-bit BCD output (4-bit tens + 4-bit ones)
);

always @* begin
    case(binary_in)
        // Tens | Ones
        5'd0:   bcd_out = 8'h00;   // 00
        5'd1:   bcd_out = 8'h01;   // 01
        5'd2:   bcd_out = 8'h02;   // 02
        5'd3:   bcd_out = 8'h03;   // 03
        5'd4:   bcd_out = 8'h04;   // 04
        5'd5:   bcd_out = 8'h05;   // 05
        5'd6:   bcd_out = 8'h06;   // 06
        5'd7:   bcd_out = 8'h07;   // 07
        5'd8:   bcd_out = 8'h08;   // 08
        5'd9:   bcd_out = 8'h09;   // 09
        5'd10:  bcd_out = 8'h10;   // 10
        5'd11:  bcd_out = 8'h11;   // 11
        5'd12:  bcd_out = 8'h12;   // 12
        5'd13:  bcd_out = 8'h13;   // 13
        5'd14:  bcd_out = 8'h14;   // 14
        5'd15:  bcd_out = 8'h15;   // 15
        5'd16:  bcd_out = 8'h16;   // 16
        5'd17:  bcd_out = 8'h17;   // 17
        5'd18:  bcd_out = 8'h18;   // 18
        5'd19:  bcd_out = 8'h19;   // 19
        5'd20:  bcd_out = 8'h20;   // 20
        5'd21:  bcd_out = 8'h21;   // 21
        5'd22:  bcd_out = 8'h22;   // 22
        5'd23:  bcd_out = 8'h23;   // 23
        5'd24:  bcd_out = 8'h24;   // 24
        5'd25:  bcd_out = 8'h25;   // 25
        5'd26:  bcd_out = 8'h26;   // 26
        5'd27:  bcd_out = 8'h27;   // 27
        5'd28:  bcd_out = 8'h28;   // 28
        5'd29:  bcd_out = 8'h29;   // 29
        5'd30:  bcd_out = 8'h30;   // 30
        5'd31:  bcd_out = 8'h31;   // 31
        default: bcd_out = 8'h00;   // Default case
    endcase
end

endmodule
'''
filename = "binary_to_bcd/binary_to_bcd.v"
# Write the extracted Verilog code to the file
with open(filename, "w") as f:
    f.write(output_verilog_code)
```
